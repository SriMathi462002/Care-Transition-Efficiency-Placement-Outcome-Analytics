a.drop('Net Flow',axis=1,inplace=True)

```python
a.columns=['Date','Inflow','Stock(CBP)','Transition','Stock(HHS)','Outflow','Net_Flow','Transfer_Efficiency']
```

```python
a.head()
```

Track Daily Movements

CBP NetChange

```python
a['CBP_NetChange']=a['Inflow']-a['Transition']
```

HHS Netchange

```python
a['HHS_NetChange']=a['Transition']-a['Outflow']
```

System Netchange

```python
a['System_NetChange']=a['Inflow']-a['Outflow']
```

```python
a.head()
```

Yesterday

```python
a['Prev_HHS'] = a['Stock(HHS)'].shift(1)
```

Apply equation

```python
a.info()
```

```python
a['Prev_HHS'] = pd.to_numeric(a['Prev_HHS'].astype(str).str.replace(',', ''), errors='coerce')
```

```python
a['Stock(HHS)'] = pd.to_numeric(a['Stock(HHS)'].astype(str).str.replace(',', ''), errors='coerce')
```

```python
a['Check'] = a['Prev_HHS'] + a['Transition'] - a['Outflow']
```

```python
a['Prev_HHS'] = a['Prev_HHS'].fillna(0)
a['Check'] = a['Check'].fillna(0)
```

```python
a['Check']
```

Compare

```python
 a['Difference'] = a['Stock(HHS)'] - a['Check']
```

```python
a.head()
```

Transition Efficiency Metrics

Transfer Efficiency Ratio

```python
a['Transfer_Efficiency_Ratio']=a['Transition']/a['Stock(CBP)']
a['Transfer_Efficiency_Ratio'].round(2)
```

Discharge Effectiveness


```python
a['Discharge_Effectiveness']=a['Outflow']/a['Stock(HHS)']
a['Discharge_Effectiveness'].round(2)
```

Pipeline Throughput Rate

Total exits ÷ Total entries

```python
a['Pipeline Throughput Rate']=a['Outflow']/a['Inflow']
a['Pipeline Throughput Rate'].round(2)
```
