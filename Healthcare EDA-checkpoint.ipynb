```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
b=pd.read_csv(r"E:\2-health_care\HHS_Unaccompanied_Alien_Children_Program_cleaned_15.csv")
```

```python
b.describe()
```

```python
b.info()
```

```python
b['Date']=pd.to_datetime(b['Date'],format='%d-%m-%Y')
```

```python
b.isnull().sum
```

```python
b.fillna(0)
```

Trend Analysis

```python
plt.figure(figsize=(12,6))
plt.plot(b['Date'],b['Inflow'],label='Inflow')
plt.plot(b['Date'],b['Outflow'],label='Outflow')
plt.legend()
plt.title('Inflow VS Outflow Over Time')
plt.show()
```

Distribution Analysis

```python
plt.figure(figsize=(12,6))
sns.histplot(b['Inflow'],kde=True,color='red',label='Inflow')
plt.title('Distribution of Inflow')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(b['Outflow'],kde=True,color='green',label='Outflow')
plt.title('Distribution of Outflow')
plt.show()
```

Correation Heatmap

```python
plt.figure(figsize=(12,6))
sns.heatmap(b.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

Rolling average of net flow

```python
b['NetFlow_MA7'] = b['Net_Flow'].rolling(7).mean()
plt.figure(figsize=(12,6))
plt.plot(b['Date'], b['Net_Flow'], label='Daily Net Flow', alpha=0.6)
plt.plot(b['Date'], b['NetFlow_MA7'], label='7-Day Rolling Avg', color='red')
plt.legend()
plt.title("Net Flow with 7-Day Rolling Average")
plt.show()
```

Scatter Plot: Throughput vs Stock(HHS)

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Stock(HHS)', y='Pipeline Throughput Rate', data=b)
plt.title("Pipeline Throughput vs Stock(HHS)")
plt.show()
```
